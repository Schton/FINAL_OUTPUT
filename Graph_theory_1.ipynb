```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a new graph
G = nx.Graph()

# Add nodes
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)

# Add edges
G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(2, 4)
G.add_edge(3, 4)

# Draw the graph
pos = nx.spring_layout(G)  # Positions for all nodes

# Draw nodes and edges
nx.draw_networkx_nodes(G, pos, node_size=700)
nx.draw_networkx_edges(G, pos, edgelist=G.edges(), width=2)
nx.draw_networkx_labels(G, pos, font_size=20, font_family="sans-serif")

# Display the plot
plt.title("Graph")
plt.show()
```

$\textbf{Graph}$

A graph $G$ consists of two finite sets: a nonempty set $V(G) of vertices and a set $E(G)$ of edges, where each edge is associated with a set consisting of either one or two vertices called its endpoints. The correspondence from edges to endpoints is called the edge-endpoint function.

$\textbf{Hence, by the definition of a graph in graph theory, this is a graph.}$

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a new directed graph
ferris_wheel_graph = nx.DiGraph()

# Define the number of compartments (vertices) in the Ferris wheel
num_compartments = 10

# Add vertices representing the compartments
for i in range(num_compartments):
    ferris_wheel_graph.add_node(i)

# Add directed edges to represent the movement
for i in range(num_compartments):
    ferris_wheel_graph.add_edge(i, (i + 1) % num_compartments)

# Draw the graph
pos = nx.circular_layout(ferris_wheel_graph)
nx.draw(ferris_wheel_graph, pos, with_labels=True, node_size=700, node_color='lightblue', font_size=10, arrowsize=20)
plt.title("Directed Graph")
plt.show()
```

$\textbf{Directed Graph}$

A directed graph, or digraph, consists of two finite sets: a nonempty set $V(G)$ of vertices and a set $D(G)$ of directed edges, where each is associated with an ordered pair of vertices called its endpoints. If edge e is associated with the pair $(v, w)$ of vertices, then e is said to be the (directed) edge from $v$ to $w$.

In this graph we can clearly see that the edges are directed to vertex, creating an ordered pair of vertices, indicating a relationship between vertices.

$\text{Hence, is a directed graph, by the definition of a directed graph, this graph satisfies the conditions of a directed graph. }$

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph object
G = nx.Graph()

# Add nodes
nodes = ['1', '2', '3', '4', '5', '6', '7']
G.add_nodes_from(nodes)

# Add edges
edges = [('1', '2'), ('1', '3'), ('2', '3'), ('4', '3'), ('5', '4'), ('6', '4'), ('5', '6')]
G.add_edges_from(edges)

# Draw the graph
nx.draw(G, with_labels=True, node_color='skyblue', font_color='black', font_weight='bold', node_size=1000, pos=nx.spring_layout(G))

# Show the graph
plt.title("Simple Graph")
plt.show()
```

$\textbf{Simple Graph}$

A simple graph is a graph that does not have any loops or parallel edges. In a simple graph, an edge with endpoints $v$ and $w$ is denoted $(v, w)$.

As you can see here, we have a disconnected graph where loops and parrallels does not exist.

$\textbf{Hence, by definition, this is a Simple graph.}$

$\textbf{Bipartite Graph}$

A bipartite graph is a graph where the vertices can be divided into two disjoint sets such that all edges connect a vertex in one set to a vertex in another set. A connection between vertices in a set does not exist.

In this graph, we have two sets, $red$ and $blue$, where there exists a connection between vertex to a vertex of another set. A connection of the same set does not exist in this graph.

$\textbf{Hence, this graph, by definition, is a Bipartite graph.}$

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()

# Add nodes and edges to the graph
G.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4), (4, 5)])

# Generate a subgraph from the original graph
sub_nodes = [1, 2, 3]
subgraph = G.subgraph(sub_nodes)

# Draw the original graph
plt.subplot(121)
nx.draw(G, with_labels=True, node_color='skyblue', font_color='black', font_weight='bold', node_size=1000)
plt.title("Original Graph")

# Draw the subgraph
plt.subplot(122)
nx.draw(subgraph, with_labels=True, node_color='lightgreen', font_color='black', font_weight='bold', node_size=1000)
plt.title("Subgraph")

# Show the plots
plt.show()
```


$\textbf{Subgraph}$

A graph $H$ is said to be a subgraph of a graph $G$ if, and only if, every vertex in $H$ is also a vertex in $G$,every edge in $H$ is also an edge in $G$, and every edge in $H$ has the same endpoints as it has in $G$.

In this generated graph, we have original graph as $G$ and subgraph as $H$, $V$ as vertices of the graph and subgraph. Where $V(G) = \{1,2,3,4,5\}$ and $V(H) = \{1,2,3\}$.

$V(G)$ contains all the vertices in $V(H)$. 

$\textbf{Hence,}$ $H$ $\textbf{is a subgraph of}$ $G.$

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph object
G = nx.Graph()

# Add nodes
nodes = ['1', '2', '3', '4', '5', '6', '7', '8']
G.add_nodes_from(nodes)

# Add edges
edges = [('1', '2'), ('3', '4'), ('1', '3'), ('2', '4'), ('1', '5'), ('2', '6') , ('3', '7') , ('4', '8') , ('5', '6') , ('7', '8') , ('5', '7') , ('6', '8')]
G.add_edges_from(edges)

# Draw the graph
nx.draw(G, with_labels=True, node_color='skyblue', font_color='black', font_weight='bold', node_size=1000, pos=nx.spring_layout(G))
plt.text(0.5, -0.3, f"Degree of vertex 4: {'3'}", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)
# Show the graph
plt.title("Degree of a vertex")
plt.show()
```

$\textbf{Degree of a vertex}$


Let $G$ be a graph and $v$ a vertex of $G$. The degree of $v$, denoted $deg(v)$, equals the number of edges that are incident on $v$, with an edge that is a loop counted twice. The total degree of $G$ is the sum of the degrees of all the vertices of $G$.

In this graph, we have to find the degree of $v(G) = \{4\}$ 

Step 1: Find the degree.

A degree of a vertex is the number of incidents the vertex has, that incident is the edge. So we have to count the number of edges connected to $v(G) = \{4\}$.

We can clearly see that there are 3.

$\textbf{Hence, the degree of vertex '4' is 3.}$

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()

# Add edges to the graph
edges = [(1, 2), (1, 3), (2, 3), (3, 4), (4, 5)]
G.add_edges_from(edges)

# Calculate the sum of degrees of all vertices
degree_sum = sum(dict(G.degree()).values())

# Calculate the number of edges
num_edges = G.number_of_edges()

# Verify the Handshake theorem
handshake_theorem_verified = (degree_sum == 2 * num_edges)

# Draw the graph
nx.draw(G, with_labels=True, node_color='skyblue', font_color='black', font_weight='bold', node_size=1000)
plt.title("The Handshake Theorem")

# Display the graph and theorem verification
plt.text(0.5, -0.1, f"Sum of degrees: {degree_sum}", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)
plt.text(0.5, -0.2, f"Number of edges: {num_edges}", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)
plt.text(0.5, -0.3, f"Handshake theorem verified: {handshake_theorem_verified}", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)
plt.show()
```

$\textbf{Handshake theorem}$

The handshake theorem states that if $G$ is any graph, then the sum of the degrees of all the vertices of $G$ equals twice the number of edges of $G$. Specifically, if the vertices of $G$ are $v_1,v_2,...,v_n, where $n$ is a non-negative integer, then the total degree of $G = deg(v_1) + deg(v_2) +...+ deg(v_n) = 2$, which is the number of edges of $G.$

The Handshaken theorem can be used to determine if a Graph exists because the total of degree should always be even.

In this graph, we represent $G$ as the graph, $v(G)$ representing vertices, $e(G)$ representing edges, and $deg(v)$ representing the degree of vertices.

Step 1: Find the number of edges and degrees.

We can see that there are exactly a total of $e(G)$ = 5 and $deg(v) = 10. 

Step 2: Verify the handshaken theorem.

Now that we have total number $e(G)$ and $deg(v)$, we can see that $deg(v)$ is exactly twice the number of $e(G)$.

$\textbf{Hence, the conditions have been satisfied and that the graph does exist, proven by Handshaken theorem.}$

$\textbf{Connected components}$

A graph $H$ is a connected component of a graph $G$ if, and only if:

1. $H$ is subgraph of $G$.

2. H is connected.

3. No connected subgraph of $G$ has $H$ as a subgraph and contains vertices or edges
that are not in $H$.

$\textbf{Euler's circuit}$

Euler's circuit uses every vertex of a graph and uses every edge exactly once. It also starts and ends with the same vertex and a graph has a Euler's circuit if and only if the graph is connected and every vertex of the graph has a positive even degree.

Suppose we started at vertex '2'.

Let us travel the graph to find if it has Euler's circuit.

$2 \rightarrow 1$

$1 \rightarrow 4$

$4 \rightarrow 5$

$5 \rightarrow 1$

$1 \rightarrow 3$

$3 \rightarrow 5$

$5 \rightarrow 6$

$6 \rightarrow 3$

$3 \rightarrow 2$

We have started and ended at the same vertex, while using every vertex atleast once and using every edge exactly once.

$\textbf{Hence, with all the satisfied conditions, this graph has a Euler's Circuit.}$

$\textbf{Euler's Trail}$

$\bullet$ Euler's circuit uses every vertex of a graph and uses every edge exactly once.

$\bullet$ If $v$ and $w$ are two distinct vertices of a graph, then there is a Euler's path from $v \rightarrow w$ if and only if the graph is connected; $v$ and $w$ have odd degree, and all the other vertices in the graph has a positive even degree.

Step 1: Identify source and destination vertices.

We identify these vertices by looking if they have odd degrees.

Suppose that vertex '4' is our source and vertex '1' is our destination.

Step 2: Traverse from source to destination.

So we do this:

$4 \rightarrow 1$

$1 \rightarrow 2$

$2 \rightarrow 3$

$3 \rightarrow 1$

Step 3: Verify Euler's trail.

So we check the conditions for Euler's trail...

Through the traversal we have:

$\bullet$ Used every vertex atleast once.

$\bullet$ Used every edge exactly once.

$\bullet$ Our source and destination vertices have odd degree, while all the other vertices have positive even degrees.

$\textbf{Hence, the graph has a Euler's trail, as all conditions have been satisfied.}$

$\bullet$ does not require all edges to be visited. However, you cannot use an edge more than once.
